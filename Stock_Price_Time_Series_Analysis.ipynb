```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ratio=np.zeros(len(df.index))
df['Return_Ratio']=ratio
```

```python
for i in range(1,len(df.index)):
    df.loc[df.index[i],'Return_Ratio']=(df.loc[df.index[i],'NIFTY50']-df.loc[df.index[i-1],'NIFTY50'])/(df.loc[df.index[i],'NIFTY50'])
df.NIFTY50=df.Return_Ratio
df.Return_Ratio[0]=0

for i in range(1,len(df.index)):
    df.loc[df.index[i],'Return_Ratio']=(df.loc[df.index[i],'TATA_STEEL']-df.loc[df.index[i-1],'TATA_STEEL'])/(df.loc[df.index[i],'TATA_STEEL'])
df.TATA_STEEL=df.Return_Ratio
df.Return_Ratio[0]=0

for i in range(1,len(df.index)):
    df.loc[df.index[i],'Return_Ratio']=(df.loc[df.index[i],'RELIANCE_INDUSTRIES']-df.loc[df.index[i-1],'RELIANCE_INDUSTRIES'])/(df.loc[df.index[i],'RELIANCE_INDUSTRIES'])
df.RELIANCE_INDUSTRIES=df.Return_Ratio
df.Return_Ratio[0]=0

for i in range(1,len(df.index)):
    df.loc[df.index[i],'Return_Ratio']=(df.loc[df.index[i],'HINDUSTAN_UNILIVER']-df.loc[df.index[i-1],'HINDUSTAN_UNILIVER'])/(df.loc[df.index[i],'HINDUSTAN_UNILIVER'])
df.HINDUSTAN_UNILIVER=df.Return_Ratio
df.Return_Ratio[0]=0

for i in range(1,len(df.index)):
    df.loc[df.index[i],'Return_Ratio']=(df.loc[df.index[i],'TATA_COMMUNICATION']-df.loc[df.index[i-1],'TATA_COMMUNICATION'])/(df.loc[df.index[i],'TATA_COMMUNICATION'])
df.TATA_COMMUNICATION=df.Return_Ratio
df.Return_Ratio[0]=0

for i in range(1,len(df.index)):
    df.loc[df.index[i],'Return_Ratio']=(df.loc[df.index[i],'FINOLEX_CABLES']-df.loc[df.index[i-1],'FINOLEX_CABLES'])/(df.loc[df.index[i],'FINOLEX_CABLES'])
df.FINOLEX_CABLES=df.Return_Ratio
df.Return_Ratio[0]=0

df.Return_Ratio=0
print(df)
```
